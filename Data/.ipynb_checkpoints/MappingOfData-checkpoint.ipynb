```python
# map or merge all postal codes to get Gemeindenummer
data_shp = df_data.merge(df_map, on='PostCode', how='left')
data_shp
```

```python
# remove unneeded columns
```

```python
data_shp = data_shp.drop(['%_IN_GDE', 'KTKZ'], axis=1)
```

```python
data_shp
```

```python
# merge 'Bezirksnummer'
```

```python
# rename the column name to be able to merge on the name
data_shp = data_shp.rename(columns={"GDENR": "BFS Gde-nummer"})
data_shp
```

```python
# review data which should be mapped
df_map_plz
```

```python
# map or merge on the field BFS Gde-nummer to get the Bezirksnummer for every municipality
data_shp = data_shp.merge(df_map_plz, on='BFS Gde-nummer', how='left')
data_shp
```

```python
# remove unneeded columns
data_shp = data_shp.drop(['GDENAMK', 'Hist.-Nummer', 'Kanton', 'Datum der Aufnahme', 'Municipality', 'Unnamed: 0'], axis=1)
```

```python
# rename some columns
data_shp = data_shp.rename(columns={"BFS Gde-nummer": "Municipality Nr.", "Gemeindename": "Municipality"})
```

```python
data_shp
```

```python
list_of_Bez = data_shp['Bezirks-nummer'].unique()
list_of_Bez
```

```python
len(list_of_Bez)
```

```python
# calculate average of Policical Orentation per Bezirk
sum_dict = {}
list_mean_Bez = []
list_Bez = []
for i in range(len(list_of_Bez)):
    list_mean_Bez.append(data_shp.loc[data_shp['Bezirks-nummer'] == list_of_Bez[i], 'Political_Value_Municipality'].mean())
    list_Bez.append(list_of_Bez[i])
    
sum_dict['Bezirks-nummer'] = list_Bez
sum_dict['Political_Value_Bezirk'] = list_mean_Bez
sum_dict
```

```python
df_to_merge = pd.DataFrame(sum_dict)
df_to_merge
```

```python
# merge Bezirksdate with data_shp
data_shp = data_shp.merge(df_to_merge, on='Bezirks-nummer', how='left')
```

```python
data_shp = data_shp.drop('Political_Value_Bezirk_x', axis=1)
```

```python
df_canton = pd.read_excel('CantonMapping.xlsx')
df_canton
```

```python
df_canton = df_canton.rename(columns={"Canton": "Canton Name",})
data_shp = data_shp.rename(columns={"Canton":"Canton Short."})
```

```python
# map canton Nr
data_shp = data_shp.merge(df_canton, on='Canton Short.', how='left')
data_shp
```

```python
# to csv
data_shp.to_csv("Agg_Bez.csv")
```
