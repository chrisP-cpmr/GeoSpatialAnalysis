```python
import pandas as pd
```

```python
elec = pd.read_csv('ch.bfe.elektrizitaetsproduktionsanlagen/ElectricityProductionPlant.csv')
```

```python
elec.head()
```

```python
cantons = (elec.Canton.unique())
cantons
```

```python
mainCat = (elec.MainCategory.unique())
mainCat
```

```python
subCat = (elec.SubCategory.unique())
subCat
```

```python
plantCat = (elec.PlantCategory.unique())
plantCat
```

```python
elec.loc[(elec['Canton'] == 'ZH') & (elec['MainCategory'] == 'maincat_2'), 'TotalPower'].sum()
```

```python
newdf = {'Canton':'', 'MainCategory':'', 'SubCategory':'', 'PlantCategory':'', 'InitialPower':0,'TotalPower':0}
df = pd.DataFrame()
df = df.append(newdf, ignore_index = True)
df
```

```python
for canton in cantons:
    for main in mainCat:
        for sub in subCat:
            for plant in plantCat:
                totalP = elec.loc[(elec['Canton'] == canton) & (elec['MainCategory'] == main) & (elec['SubCategory'] == sub) & (elec['PlantCategory'] == plant), 'TotalPower'].sum()
                initP = elec.loc[(elec['Canton'] == canton) & (elec['MainCategory'] == main) & (elec['SubCategory'] == sub) & (elec['PlantCategory'] == plant), 'InitialPower'].sum()
                newdf = {'Canton': canton, 'MainCategory':main, 'SubCategory':sub, 'PlantCategory':plant, 'InitialPower':initP,'TotalPower':totalP}
                df = df.append(newdf, ignore_index = True)
df
```
